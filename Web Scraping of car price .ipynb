```python
import selenium
import time
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys
from tqdm import tqdm

driver=webdriver.Chrome(r"C:\Users\Sony\.wdm\drivers\chromedriver\win32\102.0.5005.61\chromedriver")
driver.maximize_window()
    
cities = ['banglore','kolkata','ahmedabad','mumbai','delhi','delhi-ncr','pune','gurgaon','lucknow','jaipur','vadodara','ludhiana','chennai' , 'dehradun' , ]
name_list = []
model_list = []
amount_list = []
driven_list = []
fuel_list = []
Manufacturing = []
Location = []
for q in range(len(cities)):
    try:
        
        driver.get('https://www.cardekho.com/used-cars+in+'+str(cities[q]))
        name = driver.find_elements(By.XPATH , '//*[@class="gsc_col-md-4 gsc_col-sm-6 gsc_col-xs-12  cardnew"]//a')
        for i in range(1000):
            try:
                driver.find_element(By.XPATH,'//*[@class="gsc_col-md-4 gsc_col-sm-6 gsc_col-xs-12  cardnew"]//a').send_keys(Keys.PAGE_DOWN)
                time.sleep(.1)
            except:
                continue


        name = driver.find_elements(By.XPATH , '//*[@class="gsc_col-md-4 gsc_col-sm-6 gsc_col-xs-12  cardnew"]//a')
        model = driver.find_elements(By.XPATH , '//*[@class="car_variant"]')
        amount = driver.find_elements(By.XPATH , '//*[@class="amnt "]')
        driven = driver.find_elements(By.XPATH , '//*[@class="truncate dotlist-2"]//div[1]')
        fuel = driver.find_elements(By.XPATH , '//*[@class="truncate dotlist-2"]//div[2]') 
        locations = driver.find_element(By.XPATH , '//*[@id="rf01"]/div[1]/div/div/main/div[1]/div[2]/div/div[1]/div[1]/div[2]/input')
        location = locations.get_attribute('value')
        for i in name:
            try:
                new_name = i.text
                name_list.append(new_name[4:])
            except:
                name_list.append('-')
            try:
                year = i.text[:4]
                Manufacturing.append(year)
            except:
                Manufacturing.append('-')
        for j in model:
            model_list.append(j.text)
        for k in amount:
            amount_list.append(k.text) 
        for l in driven:
            driven_list.append(l.text)
        for m in fuel:             
            fuel_list.append(m.text)
            Location.append(str(cities[q])) 
            print(len(name_list),len(model_list),len(amount_list),len(driven_list),len(fuel_list),len(Manufacturing),len(Location))

                      
    except:
        continue
print(len(name_list),len(model_list),len(amount_list),len(driven_list),len(fuel_list),len(Manufacturing),len(Location))
df = pd.DataFrame({
    "Name":name_list,
    "Model":model_list,
    "Amount":amount_list,
    "Driven":driven_list,
    "Fuel":fuel_list,
    "Manufacturing":Manufacturing,
    "Location":Location
    }) 

excel = df.to_excel(r'C:\Users\Sony\peter_work\Used_Cardekho1.xlsx' , index=False)
df
```
